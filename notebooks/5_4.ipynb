## Chapter 5:  Exercise 5.4

Consider the truncated power series representation for cubic splines with $K$ interior knots. Let:

$$
f(X) = \sum_{j=0}^{3} \beta_j X^j + \sum_{k=i}^{K} \theta_k \left( X - \zeta_k\right)^{3}_{+}
$$

**Part 1:** Prove that the natural boundary conditions for natural cubic splines (section 5.2.1) imply the following linear constraints on the coefficients:

\begin{align}
& \beta_2 = 0, \quad \sum_{k=1}^{K} \theta_k = 0 \\
& \beta_3 = 0, \quad \sum_{k=1}^{K} \zeta_k \theta_k = 0
\end{align}

**Part 2:** Hence derive the basis (5.4) and (5.5)

---

### Part 1

Natural cubic splines constraints the function to be linear beyond the boundary knots, i.e $f''(X) = 0$ and $f'''(x) = 0$ for $X <= \zeta_1$ and $X >= \zeta_K$

\begin{align}
& f'(X) = \beta_1 + \beta_2 X + 3 \beta_3 X^2 + \sum_{k=1}^{K} 3 \theta_k (X - \zeta_k)^2 \tag{1} \\ 
& f''(X) = \beta_2 + 6 \beta_3 X + \sum_{k=1}^{K} 6 \theta_k (X - \zeta_k) \tag{2} \\
& f'''(X) =  6 \beta_3 + \sum_{k=1}^{K} 6 \theta_k \tag{3}
\end{align}

Since (2), for $X <= \zeta_1$ is equal to zero the only possible solution is that $\beta_2 = 0$ and $\beta_3 = 0$. Replacing $\beta_3 = 0$ in (3) implies that $\sum_{k=1}^{K} \theta_k = 0$. Finally replacing $\beta_2$ and $\beta_3$ in (2) for $X >= \zeta_K$:

\begin{align}
& 0 = \sum_{k=1}^{K} \theta_k (X - \zeta_k) \\
& 0 = X \sum_{k=1}^{K} \theta_k - \sum_{k=1}^{K} \theta_k \zeta_k \\
& 0 = \sum_{k=1}^{K} \theta_k \zeta_k \\
\end{align}

In conclusion: $\beta_2, \beta_3, \sum_{k=1}^{K}\theta_k, \sum_{k=1}^{K}\theta_k \zeta_k$ are all equal to 0.

---

### Part 2

Now that $\beta_2$ and $\beta_3$ are equal to 0, $f(X) = \beta_0 + \beta_1 + \sum_{k=1}^{K}\theta_k \left( X - \zeta_k \right)^{3}_{+}$. By the results from part 1 we can free two parameters, say $\theta_K$ and $\theta_{K-1}$

$$
0 = \sum_{k=1}^{K} \theta_k \Rightarrow \theta_K = - \sum_{k=1}^{K-1} \theta_k \tag{4}
$$

Combining results from part 1:

\begin{align}
\left(\sum_{k=1}^{K} \theta_k \right) \zeta_K - \sum_{k=1}^{K}\theta_k \zeta_k  & = 0 \\
\sum_{k=1}^{K} \theta_k (\zeta_K - \zeta_k)  & = 0 \\
\sum_{k=1}^{K-1} \theta_k (\zeta_K - \zeta_k)  & = 0
\end{align}

Isolationg $\theta_{K-1}$

$$
\theta_{K-1} = - \sum_{k=1}^{K-2} \frac{\theta_k (\zeta_K - \zeta_k)}{\zeta_K - \zeta_{K-1}}
$$

Replacing $\theta_K$ in $f$:

\begin{align}
f(X) & = \beta_0 + \beta_1 X + \sum_{k=1}^{K}\theta_k \left( X - \zeta_k \right)^{3}_{+} \\
& = \beta_0 + \beta_1 X + \sum_{k=1}^{K-1}\theta_k \left( X - \zeta_k \right)^{3}_{+} + \theta_K(X - \zeta_K)^{3}_{+} \\
& = \beta_0 + \beta_1 X + \sum_{k=1}^{K-1}\theta_k \left( X - \zeta_k \right)^{3}_{+} + \sum_{k=1}^{K-1} \theta_k (X - \zeta_K)^{3}_{+} \\
 & = \beta_0 + \beta_1 X + \sum_{k=1}^{K-1}\theta_k \left [\left( X - \zeta_k \right)^{3}_{+} - \left(X - \zeta_K \right)^{3}_{+} \right]  \\
 & = \beta_0 + \beta_1 X + \sum_{k=1}^{K-1}\theta_k d_k(X) \left(\zeta_K - \zeta_{k} \right)
\end{align}


Finally by replacing $\theta_{K-1}$

\begin{align}
f(X) & = \beta_0 + \beta_1 X + \sum_{k=1}^{K-2}\theta_k d_k(X) \left(\zeta_K - \zeta_{k} \right) + \theta_{K-1} d_{K-1}(X) \left(\zeta_K - \zeta_{K-1} \right)\\
& = \beta_0 + \beta_1 X + \sum_{k=1}^{K-2}\theta_k d_k(X) \left(\zeta_K - \zeta_{k} \right) -  \sum_{k=1}^{K-2} \frac{\theta_k (\zeta_K - \zeta_k)}{\zeta_K - \zeta_{K-1}} d_{K-1}(X) \left(\zeta_K - \zeta_{K-1} \right) \\
& = \beta_0 + \beta_1 X + \sum_{k=1}^{K-2}\theta_k \left(\zeta_K - \zeta_{k} \right) \left( d_k(X) -  d_{K-1}(X) \right) \\
& = \beta_0 + \beta_1 X + \sum_{k=1}^{K-2}\theta_k \left(\zeta_K - \zeta_{k} \right) N_{k+2}(X)
\end{align}

Since $(\zeta_K - \zeta_{k})$ do not depend on X, it is safe to define the new parameters $\theta'_k := \theta_k (\zeta_K - \zeta_{k})$ resulting in:

$$
f(X) = \beta_0 + \beta_1 X + \sum_{k=1}^{K-2}\theta'_k N_{k+2}(X)
$$
